```python
img_color = cv2.imread(file_list[0], cv2.IMREAD_COLOR)
```

```python
img_color.shape #이미지 배열 형태 알아보기위함
```

```python
img_color2 = cv2.imread(file_list2[0], cv2.IMREAD_COLOR) #두번째 이미지 입력
```

```python
img_color2.shape #확인
```

```python
final_value = img_color #최종 값 저장  위한 배열 선언
```

```python
final_value = np.vstack([final_value,img_color]) #vstack 사용가능한지 여부 확인
#사용 불가
```

```python
final_value.shape #확인
```

```python
#vstack 사용 불가
#np.stack 사용 가능
final_value = np.stack((final_value,img_color2), axis=0)
```

```python
final_value.shape
```

```python
final_value = np.append(final_value,img_color2) #이미 만들어진 배열에 추가 가능한지 확인
```

```python
#np.stack 으로는 불가
final_value.shape #append로 추가한 값 확인
```

```python
#실패
```

```python
#concatenate로 값 추가가능한지 확인
np.concatenate((final_value[...,np.newaxis],img_color2[...,np.newaxis]), axis=3)
```

```python
os.chdir("..") #현재위치변경
np.save('train_data.npy',final_value)#트레이닝 샘플 저장
```
