```python
import tensorflow as tf

# x_train = [1,2,3]
# y_train = [2,4,6]

X = tf.placeholder(tf.float32, shape=[None])
Y = tf.placeholder(tf.float32, shape=[None])
C = tf.placeholder(tf.float32, shape=[1])

W_plt = []
cost_plt = []


W = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]))


H = W*X+b
# testH = testW*X+b
transW = tf.assign(W,C)

cost = tf.reduce_mean(tf.square(H - Y))
# testcost = tf.reduce_mean(tf.square(testH - Y))
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.001)
train = optimizer.minimize(cost)

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in range(20001):
    cost_val,W_val,b_val,_ = sess.run([cost,W,b,train],feed_dict={X: [1,2,3,4,5], Y:[2,4,6,8,10]})
    W_plt.append(W_val)
    cost_plt.append(cost_val)
    if step % 50 == 0:
        print("step : ",step," cost : ",cost_val," W : ",W_val," b : ",b_val)
        
```

```python
#리스트에 입력한 값으로 테스트해보기
#Test from list
import matplotlib.pyplot as plt
%matplotlib inline


test_X = [6,7,8,9,10]
test_Y = [1,2,3,4,5]
X_val=sess.run(H,feed_dict={X : test_X})
plt.plot(test_X,X_val,'ro')
plt.plot([1,2,3,4,5],[2,4,6,8,10])
plt.show()
```

```python
#W값의 변화에 따른 코스트 변화
#cost values change according to W values change
plt.plot(W_plt,cost_plt,"ro")
plt.show()
```

```python
#W값의 변화에 따른 cost 변화
#cost values change according to W values change
import numpy as np

testW = np.arange(1.,3.1,0.1)
cost_test = []



for i in testW:
    sess.run(transW,feed_dict = { C : [i]})
    cost_val = sess.run(cost,feed_dict ={X:[1,2,3,4,5],Y:[2,4,6,8,10]})
    cost_test.append(cost_val)

plt.plot(testW,cost_test)
plt.show
```
